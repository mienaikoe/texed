```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./normalized_campcomp.csv', index_col='Campus Number')
```

```python
df
```

```python
df[['Campus 2016 4-Year Longitudinal: [Graduates] for All Students Rate']]
```

```python
plt.figure()

races = [
    "African American",
    "All Students",
    "American Indian",
    "Asian",
    "Hispanic",
    "Pacific Islander",
    "Two or More Races",
    "White",
]

target_cols = [
    f'Campus 2016 4-Year Longitudinal: [Graduates] for {race} Rate' for race in races
]

target_cols.append('Campus 2016 4-Year Longitudinal: Econ Disadv Denominator')

graduates_race_df = df[target_cols]
graduates_race_df.plot(x=graduates_race_df['Campus 2016 4-Year Longitudinal: Econ Disadv Denominator'])
```
